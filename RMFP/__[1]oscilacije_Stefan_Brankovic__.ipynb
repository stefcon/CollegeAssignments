### __Zadatak 1__: Posmatra se telo mase $m$ vezano za idealnu oprugu krutosti $k$, koje se kreće po glatkoj horizontalnoj podlozi duž $x$ ose.  Poznate su vrednosti sopstvene kružne učestanosti $\omega_0$, početne pozicije u kojoj se nalazi telo $x_0$ i početne brzine $v_0$.

a) Definisati funkciju __sopstvene__ koja modeluje zadati problem i omogućava rešavanje diferencijalne jednačine primenom integrate.odeint metode. Funkcija poziva parametar $\omega_0$ <font color='red'>[0.5]</font>.

```python
def sopstvene(y, t, omega0):
  y1, y2 = y
  return y2, -(omega0**2)*y1
```

b) Napisati komande koje uvoze modul numpy, modul matplotlib.pyplot i funkciju za integraciju iz scipy modula.

```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

c) Definisati listu vrednosti četiri različite sopstvene kružne učestanosti $ω_{01} = 0.5 rad/s$, $ω_{02} = 1 rad/s$, $ω_{03} = 2 rad/s$ i $ω_{04} = 4 rad/s$. <br>
Definisati početne uslove: telo u početnom trenutku miruje na rastojanju $x_0 = 5 m$ od ravnotežnog položaja.<br>
Definisati vremensku osu: 0 do 20 sekundi u 1000 ekvidistantnih tačaka. 

```python
t = np.linspace(0, 20, 1000)
omega0 = [0.5, 1, 2, 4]
pUslov = [5, 0]
```

d) Napisati kod (korišćenjem komande enumerate) koji omogućava da se u okviru istog Figure prozora (korišćenjem naredbe subplot), iscrtaju vremenski odzivi $x(t)$ za sve četiri vrednosti sopstvene učestanosti <font color='red'>[0.5]</font>. Označiti ose grafika (“vreme [s]” i “x(t) [m]”). Na graficima prikazati legendu koja se odnosi na odgovarajuću vrednost učestanosti $α$.

```python
for index, value in enumerate(omega0):
  sol = integrate.odeint(sopstvene, pUslov, t, args=(value,))
  plt.figure(1, dpi = 150)
  plt.subplot(2, 2, index + 1)
  plt.xlabel('vreme[s]')
  plt.ylabel('x(t)[m]')
  plt.subplots_adjust(hspace = .5, wspace = .3)
  plt.plot(t, sol[:,0], color = 'blue')
```

e) Napisati kod koji omogućava da se u okviru istog Figure prozora (korišćenjem naredbe subplot), iscrtaju fazni dijagrami za sve četiri vrednosti sopstvene učestanosti <font color='red'>[0.5]</font>. Označiti ose grafika (“x [m]” i “v(t) [m/s]”). Na graficima prikazati legendu koja se odnosi na odgovarajuću vrednost sopstvene učestanosti  $\omega_0$.

```python
for index, value in enumerate(omega0):
  sol = integrate.odeint(sopstvene, pUslov, t, args=(value,))
  plt.figure(2, dpi= 150)
  plt.subplot(2, 2, index + 1)
  plt.ylabel('vt(t) [m/s]')
  plt.xlabel('x[m]')
  plt.subplots_adjust(hspace = .5, wspace = .3)
  plt.plot(sol[:,0], sol[:,1], color = 'blue')

```

f) Napisati kod koji omogućava da se na istom dijagramu iscrtaju zavisnost kinetičke (plavom bojom) i potencijalne energije (crvenom bojom) sistema od vremena $t$ za $ω_0=0.5 rad/s$, ako je masa tela koje osciluje $1 kg$ <font color='red'>[0.25]</font>

```python
m = 1
omegapoc = 0.5
k = omegapoc**2 * m
sol = integrate.odeint(sopstvene, pUslov, t, args = (omegapoc,))
polozaj = sol[:,0]
brzina = sol[:,1]
plt.figure(3, figsize = (8,6), dpi = 150)
plt.plot(t, 0.5*k * polozaj**2 , label = "potenciajlna energija", color = 'red')
plt.plot(t, 0.5*m*brzina**2, label = "kineticka energija", color = 'blue')
plt.xlabel("time[s]")
plt.ylabel("Energija [J]")
plt.legend()
```

f) Napisati kod koji omogućava da se na istom dijagramu iscrta zavisnost kinetičke (plavom bojom) i potencijalne energije (crvenom bojom) sistema od koordinate $x$ <font color='red'>[0.25]</font>

```python
plt.figure(4, figsize = (8,6), dpi = 150)
plt.plot(polozaj, (polozaj**2)*k/2, label = "potencijalna energija", color = 'red' )
plt.plot(polozaj, (brzina**2)*m/2, label = "kineticka energija", color = 'blue' )
plt.xlabel("x[m]")
plt.ylabel("Energija [J]")
plt.legend()
```

f) Napisati kod koji omogućava da se iscrta zavisnost ukupne energije sistema od vremena $t$ <font color='red'>[0.25]</font>, a zatim objasniti dobijeni rezultat <font color='red'>[0.25]</font>.

```python
plt.figure(4, figsize = (8, 6), dpi = 150)
plt.plot(t, (polozaj**2)*k/2 + (brzina**2)*m/2, label = "Ukupna energija", color = 'red')
plt.xlabel("t[s]")
plt.ylabel("Ukupna energija [J]")
plt.legend()
```
